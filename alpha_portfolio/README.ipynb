# Описание идеи проекта

Это проект по ребалансировке акций в портфеле. За основу взяты сигналы из файла *101 Formulaic Alphas.pdf*

Сигналы дают числа, отрицательные и положительные. Сигнал отрицательный - стратегия предсказывает short, сигнал положительный - long. Чем больше сигнал по модулю, тем надежнее (в терминах стратегии) сигнал. 

Однако, так как наша стратегия это long only по фундаментальным факторам компаний, сигналы будут использованы только для ранжирования долей в портфеле. 

Всего у нас 3 позиции в портфеле. Акции ранжируются по сигналам, и каждой акции присваивается своя доля из списка ниже:

```python
50%
20%
15%
10%
5%
```

Самая перспективная акция (в теримнах стратегии: у которой сигнал самый большой положительный) будет иметь долю 50%, акция с самым плохим сигналом - 5%. Доли ребалансируются между торговыми днями. Для упрощения считается, что стакан всегда может проглотить объем нашей заявки. Комиссия 0,01% от объема заявки.
